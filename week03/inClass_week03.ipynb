First step toward interactivity is to make our code into a function so we can call it repeatedly!

```python
color_steps = np.linspace(0,1.0, 256) #"fake" data ranges, leave this as "global" varible
def plot_colormap(colormap_name):
    fig = plt.figure(figsize=(2,1)) # give me a figure canvas
    # first add axis on top, saving room for the colorbar on bottom
    ax = fig.add_axes([0.0, 0.15, 1.0, 0.8]) # first set of axis, left, bottom, width, height

    # now we can just use these axis like normal
    col_map = plt.get_cmap(colormap_name)
    colors = col_map(color_steps)[:,:3] # RGB channels, ignoring any Alpha=opacity channel
    ax.plot(color_steps, colors[:,0], '-r')
    ax.plot(color_steps, colors[:,1], '-g')
    ax.plot(color_steps, colors[:,2], '-b')
    # let's turn off the axis labels
    ax.xaxis.set_visible(False)
    ax.yaxis.set_visible(False)
    # no tick labels
    ax.xaxis.set_ticklabels([])
    ax.yaxis.set_ticklabels([])
    # also set the x/y axis scales explicity
    ax.set_xlim(0,1)
    ax.set_ylim(0,1.01)

    # now adding colorbar plot on bottom
    ax_colorbar = fig.add_axes([0.0, 0.0, 1.0, 0.2])
    matplotlib.colorbar.ColorbarBase(ax_colorbar, cmap=col_map, orientation='horizontal')
    ax_colorbar.xaxis.set_visible(False)
    plt.show()
```

Finally, let's make this interactive with ipywidgets:

```python
color_steps = np.linspace(0,1.0, 256) #"fake" data ranges, leave this as "global" varible

@ipywidgets.interact(colormap_name=plt.colormaps())
def plot_colormap(colormap_name):
    fig = plt.figure(figsize=(2,1)) # give me a figure canvas
    # first add axis on top, saving room for the colorbar on bottom
    ax = fig.add_axes([0.0, 0.15, 1.0, 0.8]) # first set of axis, left, bottom, width, height

    # now we can just use these axis like normal
    col_map = plt.get_cmap(colormap_name)
    colors = col_map(color_steps)[:,:3] # RGB channels, ignoring any Alpha=opacity channel
    ax.plot(color_steps, colors[:,0], '-r')
    ax.plot(color_steps, colors[:,1], '-g')
    ax.plot(color_steps, colors[:,2], '-b')
    # let's turn off the axis labels
    ax.xaxis.set_visible(False)
    ax.yaxis.set_visible(False)
    # no tick labels
    ax.xaxis.set_ticklabels([])
    ax.yaxis.set_ticklabels([])
    # also set the x/y axis scales explicity
    ax.set_xlim(0,1)
    ax.set_ylim(0,1.01)

    # now adding colorbar plot on bottom
    ax_colorbar = fig.add_axes([0.0, 0.0, 1.0, 0.2])
    matplotlib.colorbar.ColorbarBase(ax_colorbar, cmap=col_map, orientation='horizontal')
    ax_colorbar.xaxis.set_visible(False)
    plt.show()
```
